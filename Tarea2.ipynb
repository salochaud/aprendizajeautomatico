# Tarea 2
Este conjunto de datos contiene información sobre los miembros de un gimnasio, incluyendo variables como la edad, sexo, peso, altura, tipo de membresía, frecuencia semanal de asistencia, progreso físico y duración de la membresía.

Datos obtenidos de Kaggle – Gym Members Exercise Dataset.



## Carga de conjunto de datos

```python
import pandas as pd

df = pd.read_csv("/content/gym_members_exercise_tracking.csv")

display(df.head())
```

## Tipo de dato por columna

```python
df.dtypes
```

## Se asigna una columna como indice

```python
df = df.set_index("Gender")
display(df.head())
```

## Crear, modificar o eliminar columnas

```python

df["BMI"] = df["Weight (kg)"] / ((df["Height (m)"]/100)**2)

df["Low_Progress"] = df["Fat_Percentage"] < 50
display(df.head())
```

## Filtrar registros

```python
df_filtered = df.query("BMI > 25 and Fat_Percentage <= 30")
display(df_filtered)
```

## Se creo un subconjunto de datos

```python
Strenght = df[df["Workout_Type"] == "Strength"]
display(Strenght.head())
```

## Se creo un grupo de datos

```python
avg_bmi_gender = df.groupby("Gender")["BMI"].mean()
display(avg_bmi_gender)
```
